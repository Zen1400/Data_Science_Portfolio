# Chi Squared Test



- Used with count data to test how well it fits some expected distribution. The most common use of the chi-square statistic in statistical practice is with r  c contingency tables, to assess whether the null hypothesis of independence among variables is reasonable


- A measure of the extent to which some observed data departs from expectation.

## Test 3 Headlines, run each on 1000 visitors




Note : The overall number is 34 / 3000 which means if the Null Hyp is true for 1000 we expect 11.33





### Degree of Freedom



```python
import numpy as np
from scipy.stats import chi2_contingency


observed_table = np.array([[14, 8, 12],
                           [986, 992, 988]])

# Perform the Chi-Square test
chi2, p, _, _ = chi2_contingency(observed_table)

# p is the p-value
print("Chi-Square Statistic:", chi2)
print("P-value:", p)
```
